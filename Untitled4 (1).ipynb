# Data Prepration

```python
import os
import glob
import cv2
import numpy as np

# --- Configuration ---
IMG_SIZE = (224, 224)
DATASET_DIR = "./cp_dataset"  # The folder where we extracted the RAR file

# ==========================================
# PHASE 1 & 2: DATA EXTRACTION & LOADING
# ==========================================
def load_cp_anemic_dataset(dataset_path):
    print("--- Phase 2: Loading CP-AnemiC Images (CIELAB + CLAHE) ---")
    images, labels = [], []

    # Note: Changed 'Healthy' to 'Non-anemic' to match the rar file folders
    classes = {'Anemic': 1.0, 'Non-anemic': 0.0}

    # Initialize CLAHE object for contrast enhancement
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

    for class_name, label in classes.items():
        folder_path = os.path.join(dataset_path, class_name)
        img_paths = glob.glob(os.path.join(folder_path, "*.*"))
        print(f"Processing {len(img_paths)} images in '{class_name}'...")

        for img_path in img_paths:
            try:
                # 1. Read Image
                img = cv2.imread(img_path)
                if img is None: continue

                # 2. Convert BGR to CIELAB
                lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

                # 3. Apply CLAHE to the L-channel (Lightness)
                l, a, b = cv2.split(lab)
                l_enhanced = clahe.apply(l)

                # 4. Merge back and Resize
                lab_enhanced = cv2.merge((l_enhanced, a, b))
                img_resized = cv2.resize(lab_enhanced, IMG_SIZE)

                images.append(img_resized)
                labels.append(label)
            except Exception as e:
                print(f"Error processing {img_path}: {e}")

    return np.array(images, dtype=np.float32), np.array(labels, dtype=np.float32)
```

# Model Prepration


```python
#RGB Intensity Model
class RGBIntensityModel:
    def __init__(self):
        self.threshold = 0.0
        self.channel = 0 # 0=Red, 1=Green, 2=Blue

    def fit(self, X, y):
        # We calculate the mean intensity of the RED channel for each image
        # X shape: (num_images, 224, 224, 3)
        red_intensities = np.mean(X[:, :, :, self.channel], axis=(1, 2))

        best_acc = 0
        best_t = 0

        # Search for the best threshold between the min and max intensity found
        thresholds = np.linspace(red_intensities.min(), red_intensities.max(), 100)

        for t in thresholds:
            # Prediction: If intensity > threshold, it's 'Paler' (Anemic = 1.0)
            # Note: Depending on your lighting, it could be reds < t.
            # We check both directions.
            preds = (red_intensities > t).astype(float)
            acc = accuracy_score(y, preds)

            if acc < 0.5: # If accuracy is low, try the inverse relationship
                preds = (red_intensities < t).astype(float)
                acc = accuracy_score(y, preds)

            if acc > best_acc:
                best_acc = acc
                best_t = t

        self.threshold = best_t
        print(f"Optimal RGB Intensity Threshold found: {self.threshold:.4f}")

    def predict(self, X):
        red_intensities = np.mean(X[:, :, :, self.channel], axis=(1, 2))
        # Apply the learned threshold
        return (red_intensities > self.threshold).astype(float)
```

```python
#Resnet Model
import tensorflow as tf
from tensorflow.keras import layers, models

class SEBlock(layers.Layer):
    """Squeeze-and-Excitation Block for feature recalibration """
    def __init__(self, ratio=16, **kwargs):
        super(SEBlock, self).__init__(**kwargs)
        self.ratio = ratio

    def build(self, input_shape):
        self.channels = input_shape[-1]
        self.squeeze = layers.GlobalAveragePooling2D()
        self.excitation = models.Sequential([
            layers.Dense(self.channels // self.ratio, activation='relu'),
            layers.Dense(self.channels, activation='sigmoid')
        ])
        super(SEBlock, self).build(input_shape)

    def call(self, inputs):
        squeeze = self.squeeze(inputs)
        excitation = self.excitation(squeeze)
        return inputs * tf.reshape(excitation, [-1, 1, 1, self.channels])

def build_resnet_anemia_model(learning_rate=3e-5, dense_units=256, dropout=0.4):
    """
    Builds a ResNet50 model using parameters optimized for deep learning
    and anemia detection.
    """
    # Initialize ResNet50 with ImageNet weights
    base_model = tf.keras.applications.ResNet50(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )

    # Freeze the base to protect pre-trained weights during initial training
    base_model.trainable = False

    inputs = layers.Input(shape=(224, 224, 3))
    x = base_model(inputs, training=False)

    # Add SE-Block for attention on relevant color channels
    x = SEBlock(ratio=16)(x)

    # Global Pooling as recommended in the ResNet architecture
    x = layers.GlobalAveragePooling2D()(x)

    # Optimized Classification Head
    x = layers.Dense(dense_units, activation='relu')(x)
    x = layers.BatchNormalization()(x) # Ensures healthy signal variance
    x = layers.Dropout(dropout)(x)

    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs, outputs)

    # Use Adam optimizer with a low learning rate for stability
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )

    return model
```

```python
#Densenet Model
import tensorflow as tf
from tensorflow.keras import layers, models

class SEBlock(layers.Layer):
    """Squeeze-and-Excitation Block for feature recalibration"""
    def __init__(self, ratio=16, **kwargs):
        super(SEBlock, self).__init__(**kwargs)
        self.ratio = ratio

    def build(self, input_shape):
        self.channels = input_shape[-1]
        self.squeeze = layers.GlobalAveragePooling2D()
        self.excitation = models.Sequential([
            layers.Dense(self.channels // self.ratio, activation='relu'),
            layers.Dense(self.channels, activation='sigmoid')
        ])
        super(SEBlock, self).build(input_shape)

    def call(self, inputs):
        squeeze = self.squeeze(inputs)
        excitation = self.excitation(squeeze)
        return inputs * tf.reshape(excitation, [-1, 1, 1, self.channels])

def build_densenet_anemia_model(learning_rate=3e-5, dense_units=256, dropout=0.3):
    """
    Builds a DenseNet121 model tailored for Anemia Detection.
    """
    # Initialize DenseNet121 with ImageNet weights
    base_model = tf.keras.applications.DenseNet121(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )

    # Freeze the base to protect pre-trained weights during the initial phase
    base_model.trainable = False

    inputs = layers.Input(shape=(224, 224, 3))
    x = base_model(inputs, training=False)

    # Add SE-Block for attention on relevant feature maps
    x = SEBlock(ratio=16)(x)

    # Global Pooling
    x = layers.GlobalAveragePooling2D()(x)

    # Classification Head
    x = layers.Dense(dense_units, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(dropout)(x)

    # Binary output (Anemic vs Healthy)
    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs, outputs)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )

    return model
```

# Model Training

```python
import time
import psutil
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score

if __name__ == "__main__":
    # 1. Load Data directly using the new CP-AnemiC function (No prep_indian_dataset needed)
    X, y = load_cp_anemic_dataset(DATASET_DIR)

    if len(X) > 0:
        print(f"\nSuccessfully loaded {len(X)} images total.")
        print("\n--- Phase 3: K-Fold Training & Evaluation ---")
        X = X / 255.0  # Normalize to [0, 1]

        # 2. Setup 5-Fold Cross-Validation
        kf = KFold(n_splits=5, shuffle=True, random_state=42)
        all_final_results = []

        models_to_test = ['LAB_Intensity_Baseline', 'ResNet50_SE', 'DenseNet121_SE']

        for m_name in models_to_test:
            print(f"\n{'='*50}")
            print(f"Evaluating {m_name} with 5-Fold CV")
            print(f"{'='*50}")

            fold_metrics = []

            for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
                print(f"\n--- Training Fold {fold + 1} ---")
                X_train, X_val = X[train_idx], X[val_idx]
                y_train, y_val = y[train_idx], y[val_idx]

                proc = psutil.Process()
                proc.cpu_percent(None) # Reset CPU counter
                start_time = time.time()

                # --- MODEL SELECTION & TRAINING ---
                if m_name == 'LAB_Intensity_Baseline':
                    model = RGBIntensityModel()
                    model.channel = 1  # IMPORTANT: Set to 1 for the 'a' (red/green) channel in LAB space
                    model.fit(X_train, y_train)
                    preds = model.predict(X_val)

                    # Calculate AUC for baseline (discrete predictions)
                    try: auc = roc_auc_score(y_val, preds)
                    except ValueError: auc = 0.5

                else:
                    tf.keras.backend.clear_session()
                    if m_name == 'ResNet50_SE':
                        model = build_resnet_anemia_model(learning_rate=3e-5)
                    elif m_name == 'DenseNet121_SE':
                        model = build_densenet_anemia_model(learning_rate=3e-5)

                    # Advanced Callbacks
                    callbacks = [
                        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
                        tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)
                    ]

                    # Train Deep Learning Model
                    model.fit(
                        X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=15,
                        batch_size=16,
                        callbacks=callbacks,
                        verbose=0
                    )

                    # Predict
                    preds_proba = model.predict(X_val, verbose=0).flatten()
                    preds = (preds_proba > 0.5).astype(float)
                    try: auc = roc_auc_score(y_val, preds_proba)
                    except ValueError: auc = 0.5

                # --- METRICS CALCULATION ---
                end_time = time.time()
                cpu_util = proc.cpu_percent(None)

                acc = accuracy_score(y_val, preds)
                rec = recall_score(y_val, preds, zero_division=0)
                inf_time = ((end_time - start_time) / len(X_val)) * 1000 # milliseconds per image

                print(f"Fold {fold+1} Results -> Acc: {acc:.4f} | AUC: {auc:.4f} | Recall: {rec:.4f}")

                fold_metrics.append({
                    "Accuracy": acc,
                    "AUC": auc,
                    "Recall": rec,
                    "Inference Time (ms)": inf_time,
                    "CPU (%)": cpu_util
                })

            # --- AGGREGATE FOLD RESULTS ---
            avg_acc = np.mean([m["Accuracy"] for m in fold_metrics])
            avg_auc = np.mean([m["AUC"] for m in fold_metrics])
            avg_rec = np.mean([m["Recall"] for m in fold_metrics])
            avg_inf = np.mean([m["Inference Time (ms)"] for m in fold_metrics])
            avg_cpu = np.mean([m["CPU (%)"] for m in fold_metrics])

            all_final_results.append({
                "Model": m_name,
                "Avg Accuracy": round(avg_acc, 4),
                "Avg AUC": round(avg_auc, 4),
                "Avg Recall": round(avg_rec, 4),
                "Inference Time (ms)": round(avg_inf, 2),
                "CPU Util (%)": round(avg_cpu, 2)
            })

        # 3. Final Output
        df_results = pd.DataFrame(all_final_results)
        print("\n" + "="*70)
        print("FINAL 5-FOLD CROSS-VALIDATION COMPARISON")
        print("="*70)
        print(df_results.to_string(index=False))
        print("="*70)

    else:
        print("Error: Dataset empty after loading. Check the folders.")
```
